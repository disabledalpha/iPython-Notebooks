## Auxiliary functions

```python
def rgb2gray(rgb):
    r, g, b = rgb[:,:,0], rgb[:,:,1], rgb[:,:,2]
    gray = 0.2989 * r + 0.5870 * g + 0.1140 * b
    return gray
```

```python
def generate_square(size_of_image=(256,256),size_of_square=(100,100)):
    size_of_image=(256,256)
    size_of_square=(100,100)
    image = np.zeros(size_of_image,dtype=float)
    center = (list(size_of_image)+np.array([1,1]))/2
    left_up_angle = center-(list(size_of_square)+np.array([1,1]))/2
    for i in range(size_of_square[0]):
        for j in range(size_of_square[1]):
            image[left_up_angle[0]+i,left_up_angle[1]+j]=1.
    return image
```

```python
# works only with square images!
def enlarge_image(original, final_dim):
    diff = final_dim - original.shape[0]
    if diff<=0:
        print("final dimension smaller or equal than dimension of original image")
    #enlarge
    enlarged = np.append(np.append(original,np.zeros((original.shape[0],diff)), axis=1),np.zeros((diff,final_dim)),axis=0)
    #center
    result = np.roll(np.roll(enlarged,diff/2,axis=0),diff/2,axis=1)
    return result
```

## Test with squares

```python
# A black square
square = generate_square()
plt.imshow(square, cmap = cm.binary)
```

```python
# The above square rotated of 60deg (and thus enlarged)
square60 = ndimage.interpolation.rotate(square,60, prefilter = False)
plt.imshow(square60, cmap = cm.binary)
plt.colorbar()
```

```python
# Percentage of non-zero elements in the turned square
float(np.count_nonzero(square60))/(350*350)
```

```python
# We enlarge also the first square, for comparison
square = enlarge_image(square,square60.shape[0])
plt.imshow(square, cmap = cm.binary)
```

```python
# Percentage of non-zero elements in the originila (enlarged) square
float(np.count_nonzero(square))/(350*350)
```

```python
# And we rotate back the rotated square
square_returned = ndimage.interpolation.rotate(square60,-60,reshape=False, prefilter=False)
plt.imshow(square_returned, cmap= cm.binary)
```

```python
# Percentage of non-zero elements in the returned square
float(np.count_nonzero(square_returned))/(350*350)
```

```python
# difference between the two squares
plt.imshow(abs(square-square_returned), cmap = cm.binary)
plt.colorbar()
```

```python
# Percentage of pixels that do not coincide in the two images
float(np.count_nonzero(square-square_returned))/(350**2)
```

*By turning and turning back we do not recover the original image!*

### FFTs

In all the graphs below we plot the logarithm of the absolute values of the FFTs (Vincente's suggestion).

```python
# FFT of the non-rotated enlarged square
square_fft = np.fft.fftshift(np.fft.fft2(square))
plt.imshow(np.log(abs(square_fft)))
```

```python
# FFT of the rotated square
square60_fft = np.fft.fftshift(np.fft.fft2(square60))
plt.imshow(np.log(abs(square60_fft)))
```

```python
# FFT of the re-rotated square (which should 
# coincide with the rotated back FFT of the 
# rotated square)
square_returned_fft = np.fft.fftshift(np.fft.fft2(square_returned))
plt.imshow(np.log(abs(square_returned_fft)))
```

```python
# Difference between the FFTs
diff = square_returned_fft - square_fft
plt.imshow(np.log(abs(diff)))
```

```python
# Norms of the FFTs
print("Norm of original image FFT: {}").format(LA.norm(square_fft))
print("Norm of turned image FFT: {}").format(LA.norm(square_returned_fft))
print("Norm of the difference of the FFTs: {}").format(LA.norm(square_fft-square_returned_fft))
print("Relative error w.r.t. the norm of the original image: {0:.2f}%").format(100*LA.norm(square_fft-square_returned_fft)/LA.norm(square_fft))
```

### Fixed vs turned/returned

```python
def test_FFT(image, angle = 60):
    if image.shape[0]!=image.shape[1]:
        print("Only square images!")
        return
    
    image_turned = ndimage.interpolation.rotate(image, angle, prefilter = False)
    image_enlarged = enlarge_image(image,image_turned.shape[0])
    image_returned = ndimage.interpolation.rotate(image_turned,-angle,reshape=False, prefilter=False)
    
    image_fft = np.fft.fftshift(np.fft.fft2(image_enlarged))
    image_returned_fft = np.fft.fftshift(np.fft.fft2(image_returned))
    
    orig_norm = LA.norm(image_fft)
    returned_norm = LA.norm(image_returned_fft)
    diff_norm = LA.norm(image_fft-image_returned_fft)
    
    print("Norm of original image FFT: {}").format(orig_norm)
    print("Norm of turned image FFT: {}").format(returned_norm)
    print("Norm of the difference of the FFTs: {}").format(diff_norm)
    print("Relative error w.r.t. the norm of the original image: {0:.2f}%").format(100*diff_norm/orig_norm)
    print("Norm of the difference of the normalized FFTs: {}").format(LA.norm(image_fft/orig_norm - image_returned_fft/returned_norm))
```

```python
def test_FFT_results(image, angle=60):
    if image.shape[0]!=image.shape[1]:
        print("Only square images!")
        return
    
    image_turned = ndimage.interpolation.rotate(image, angle, prefilter = False)
    image_enlarged = enlarge_image(image,image_turned.shape[0])
    image_returned = ndimage.interpolation.rotate(image_turned,-angle,reshape=False, prefilter=False)
    
    image_fft = np.fft.fftshift(np.fft.fft2(image_enlarged))
    image_returned_fft = np.fft.fftshift(np.fft.fft2(image_returned))
    
    image_fft_normalized = image_fft/ LA.norm(image_fft)
    image_returned_fft_normalized = image_returned_fft/ LA.norm(image_returned_fft)
    diff_norm = LA.norm(image_fft_normalized-image_returned_fft_normalized)
    
    return diff_norm
```

```python
def test_FFT_results_not_normalized(image, angle=60):
    if image.shape[0]!=image.shape[1]:
        print("Only square images!")
        return
    
    image_turned = ndimage.interpolation.rotate(image, angle, prefilter = False)
    image_enlarged = enlarge_image(image,image_turned.shape[0])
    image_returned = ndimage.interpolation.rotate(image_turned,-angle,reshape=False, prefilter=False)
    
    image_fft = np.fft.fftshift(np.fft.fft2(image_enlarged))
    image_returned_fft = np.fft.fftshift(np.fft.fft2(image_returned))
    
    diff_norm = LA.norm(image_fft-image_returned_fft)
    
    return diff_norm
```

```python
def test_folder(folder_name=folder):
    for filename in os.listdir(folder_name):
        infilename = os.path.join(folder_name, filename)
        if not os.path.isfile(infilename): 
            continue

        base, extension = os.path.splitext(infilename)
        if extension == '.png':
            test_img = 1 - rgb2gray(plt.imread(infilename))
#             plt.imshow(test_img, cmap = cm.binary)
            print("**IMAGE:** {}").format(base)
            test_FFT(test_img)
            print("\n")
```

### Difference between the FFTs of different images

```python
def crossed_test(image,folder_name=folder):
    if image.shape[0]!=image.shape[1]:
        print("Only square images!")
        return
    
    image_fft = np.fft.fftshift(np.fft.fft2(image))
    image_norm = LA.norm(image_fft)
    image_fft_normalized = image_fft / image_norm
    
    for filename in os.listdir(folder_name):
        infilename = os.path.join(folder_name, filename)
        if not os.path.isfile(infilename): 
            continue

        base, extension = os.path.splitext(infilename)
        if extension == '.png':
            test_img = 1 - rgb2gray(plt.imread(infilename))
            print("**IMAGE:** {}").format(base)
            
            if test_img.shape!= image.shape:
                print("Test images must have the same dimension as original image!")
                print("\n")
                continue
            
            test_fft = np.fft.fftshift(np.fft.fft2(test_img))
            test_norm = LA.norm(test_fft)
            diff_norm = LA.norm(test_fft-image_fft)
            print("Norm of the difference of the FFTs: {}").format(diff_norm)
            print("Norm of the difference relative to the norm of the fixed image: {0:.2f}%").format(100*diff_norm/image_norm)
                
            test_fft_normalized = test_fft / test_norm
            diff_norm = LA.norm(test_fft_normalized-image_fft_normalized)
            print("Norm of the difference of the normalized FFTs: {}").format(diff_norm)
            
            print("\n")
```

```python
def crossed_test_array(image,folder_name=folder):
    if image.shape[0]!=image.shape[1]:
        print("Only square images!")
        return
    
    image_fft = np.fft.fftshift(np.fft.fft2(image))
    image_norm = LA.norm(image_fft)
    image_fft_normalized = image_fft / image_norm

    results = []
    
    for filename in os.listdir(folder_name):
        infilename = os.path.join(folder_name, filename)
        if not os.path.isfile(infilename): 
            continue
            
        base, extension = os.path.splitext(infilename)
        if extension == '.png':            
            test_img = 1 - rgb2gray(plt.imread(infilename))
            
            if test_img.shape!= image.shape:
                continue
            
            test_fft = np.fft.fftshift(np.fft.fft2(test_img))
            test_norm = LA.norm(test_fft)
            test_fft_normalized = test_fft / test_norm
            diff_norm = LA.norm(test_fft_normalized-image_fft_normalized)
            
            if results == []:
                results = [os.path.splitext(filename)[0],diff_norm]
            else:
                results = np.vstack([results,[os.path.splitext(filename)[0],diff_norm]])
    
    for i in range(1,6):
        results = np.vstack([results,[str(i*30)+'returned', test_FFT_results(image, 30*i)]])
    
    return results
```

### Graphs

Here we graph the differences between the normalized FFTs of the images in the folder

```python
def plot_comparison(image):
    comparison = crossed_test_array(image)
    plt.plot(comparison[:,1], 'ro')
    for i in range(len(comparison[:,1])):
        plt.text(i,comparison[i,1],comparison[i,0][:3] )
```
